```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../static/data/bAllSpeedData.csv')
```

```python
df = df.transpose()
```

```python
len(df.columns[0:12])
```

```python
# 12개의 컬럼씩 합산하여 새로운 데이터프레임 생성
merged = pd.DataFrame()
merged.index = df[0].index
for i in range(0, len(df.columns), 12):
    cols = df.columns[i:i+12]
    sums = df[cols].sum(axis=1) / len(cols)
    sums.columns[0] = i / 12
    merged = pd.concat([merged, sums], axis=1)
```
